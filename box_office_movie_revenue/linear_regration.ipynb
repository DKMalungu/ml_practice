```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Reading CSV file
df = pd.read_csv("./data/cost_revenue_clean.csv")
```

```python
# removing excess columns
df = df[['production_budget_usd', 'worldwide_gross_usd']]
```

```python
# Sample Data
df.sample(n=10)
```

```python
# Data description
df.describe()
df.columns
```

```python
# creating feature matrix and target datasets
# feature matrix
X = pd.DataFrame(data=df, columns=["production_budget_usd"], dtype=np.float32)
# target
y = pd.DataFrame(data=df, columns=["worldwide_gross_usd"],  dtype=np.float32)
```

```python
type(X)
```

```python
# Visualizing using a scatter plot
plt.figure(figsize=(10,6))
plt.scatter(x=X, y=y, alpha=0.5)
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.xlabel("Movie Production Budget in USD($)", color='yellow')
plt.ylabel("Movie WorldWide Gross in USD($)", color='yellow')
plt.title("A Graph of Movie Production Budget against World Wide Gross in USD($)", color='yellow')
plt.show()
```

```python
# Training The model to make some predictions
model_regression = LinearRegression()
model_regression.fit(X, y)
```

```python
# Visualizing the Slope cofe
model_regression.coef_
```

```python
# Visualizing the y intercept
model_regression.intercept_
```

```python
# Visualizing the scatter plot with the line of best fit
plt.figure(figsize=(10, 6))
plt.scatter(x=X, y=y, alpha=0.5)
plt.plot(X, model_regression.predict(X), color='Red', linewidth=4)
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.xlabel("Movie Production Budget in USD($)", color='yellow')
plt.ylabel("Movie WorldWide Gross in USD($)", color='yellow')
plt.title("A Graph of Movie Production Budget against World Wide Gross in USD($)", color='yellow')
plt.show()
```

```python
# Model Coverage
model_regression.score(X, y)*100
```
