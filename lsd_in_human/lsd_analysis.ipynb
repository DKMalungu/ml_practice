```python
# Important Libraries
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# importing data
raw_data_path = 'data/lsd_math_score_data.csv'
data = pd.read_csv(raw_data_path)
```

```python
# Displaying data
data
```

```python
# Descriptive analysis
data.describe()
```

```python
time = pd.DataFrame(data, columns=['Time_Delay_in_Minutes'])
lsd = pd.DataFrame(data, columns=['LSD_ppm'])
score = pd.DataFrame(data, columns=['Avg_Math_Test_Score'])
```

```python
plt.figure(figsize=(10,6))
plt.plot(time, lsd, linewidth=4, color='b')
plt.plot(time, score, alpha=0.5, color='g')
plt.xlabel("Time in minutes", fontsize=28)
# plt.xlim(0, 500)
plt.yticks(fontsize=17)
plt.ylabel("LSD in PPM", fontsize=28)
plt.yticks(fontsize=17)
# plt.ylim(0, 6.5)
plt.title("Time vs LSD", fontsize=28)
plt.style.use("dark_background")
plt.show()
```

```python
# Creating Linear Regression Model Object
regression_model = LinearRegression()
```

```python
# Training Models
regression_model.fit(lsd, score)
```

```python
# Y intercept
y_intercept = regression_model.intercept_[0]
print("The Y Intercept (Intercept): ", y_intercept)
# Gradiant
g_value = regression_model.coef_[0][0]
print("Gradient (Theta): ", g_value)
r_square = regression_model.score(lsd, score)
print("R-Square: ", r_square)
```

```python
# plot of actual values vs predicted values
plt.figure(figsize=(10,6))
plt.scatter(x=lsd, y=score, s = 200, alpha=0.5, color='b')
plt.plot(lsd, regression_model.predict(lsd), linewidth=4, color='g')
plt.xlabel("LSD in PPM", fontsize=28)
# plt.xlim(0, 500)
plt.yticks(fontsize=17)
plt.ylabel("Score", fontsize=28)
plt.yticks(fontsize=17)
# plt.ylim(0, 6.5)
plt.title("LSD vs Score", fontsize=28)
plt.style.use("dark_background")
plt.show()
```
