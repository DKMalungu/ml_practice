### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Example 1
$$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

### Generating data using numpy

```python
x_1 = np.linspace(start=-3, stop=3, num=50000)
print(x_1)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_1, f(x_1), linewidth=4)
plt.xlim(-3, 3)
plt.xticks(fontsize=17)
plt.xlabel("x", fontsize=17)
plt.ylim(0, 10)
plt.yticks(fontsize=17)
plt.ylabel("f(x)", fontsize=17)
plt.style.use('dark_background')
plt.show()
```
