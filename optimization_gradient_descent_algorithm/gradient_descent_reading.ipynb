### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Example 1
$$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

### Generating data using numpy

```python
x_1 = np.linspace(start=-3, stop=3, num=50000)
print(x_1)
```

```python
def df(x):
    return 2*x+1
```

#### Derivative and Slope

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_1, f(x_1), linewidth=4)
plt.plot(x_1, df(x_1), color='b')
plt.xlim(-3, 3)
plt.xticks(fontsize=17)
plt.xlabel("x", fontsize=17)
plt.ylim(0, 10)
plt.yticks(fontsize=17)
plt.ylabel("f(x)", fontsize=17)
plt.title("Graph for f(x) = x^2+x+1 for values between -3 and 3", fontsize=17)
plt.style.use('dark_background')
plt.show()
```

```python
# First Method
plt.figure(figsize=[10, 10])

# COST Function
# (Row, Column, Position)
plt.subplot(2, 1, 1)
plt.plot(x_1, f(x_1), linewidth=4, color='b')
plt.title("Graph for f(x) = x^2+x+1 and Derivative of ")
plt.ylabel("f(x)")
plt.xlabel("X")

plt.subplot(2, 1, 2)
plt.plot(x_1, df(x_1), linewidth=4, color='g')
plt.ylabel("Derivative of f(x)")
plt.xlabel("X")
plt.style.use("dark_background")
plt.show()
```

```python
# Second Method
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=[10,10])
fig.suptitle("Graph for f(x) = x^2+x+1 for values between -3 and 3")

ax1.plot(x_1, f(x_1), linewidth=4)
ax1.set_ylabel("f(X)")
ax1.set_xlabel("X Values")

ax2.plot(x_1, df(x_1), linewidth=4)
ax2.set_ylabel("Derivative of f(x)")
ax2.set_xlabel("X Values")
plt.style.use("dark_background")
plt.show()
```
